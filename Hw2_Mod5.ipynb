```python
df = pd.DataFrame(python_data)
# Функція для розрахунку середнього значення зарплати на місяць
# Функція для розрахунку середнього значення зарплати на місяць
def fill_avg_salary(salary: int) -> float:
    return salary // 12  # Ділимо націло, оскільки цілий рік
df['G'] = df['G'].astype(int)
# Застосовуємо функцію до кожного рядка за допомогою методу apply
df['avg'] = df['G'].apply(fill_avg_salary).astype(int)

print(df)
```
